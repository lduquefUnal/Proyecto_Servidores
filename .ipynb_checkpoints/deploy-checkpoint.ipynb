```python
import boto3, sagemaker
from sagemaker import Session, image_uris
from sagemaker.multidatamodel import MultiDataModel
from sagemaker.sklearn.model import SKLearnModel

region = boto3.Session().region_name
role = sagemaker.get_execution_role()  # o sagemaker.get_execution_role() si estás en notebook de SM

image_uri = image_uris.retrieve(framework="sklearn", region=region, version="1.7-1")
model_prefix = "s3://modelos-sagemaker/modelos/sentimientos/"  # carpeta donde están los tar.gz

# Modelo base con tu inference.py (el mismo que va dentro de los tar)
base_model = SKLearnModel(
    model_data=None,
    role=role,
    entry_point="inference.py",
    source_dir="code",   # la carpeta local con inference.py
    image_uri=image_uri,
)

mme = MultiDataModel(
    name="sentiment-mme",
    model_data_prefix=model_prefix,
    model=base_model,
    role=role,
    sagemaker_session=Session(),
)

# Registrar ambos tar
mme.add_model(f"{model_prefix}svm_tfidfvectorizer/model_svm_tfidfvectorizer.tar.gz")
mme.add_model(f"{model_prefix}svm_countvectorizer/model_svm_countvectorizer.tar.gz")

predictor = mme.deploy(
    initial_instance_count=1,
    instance_type="ml.m5.large",
    endpoint_name="sentiment-mme-endpoint",
)

# Invocación: selecciona el tar con target_model
resp_tfidf = predictor.predict(
    {"input": ["el servicio fue excelente"]},
    target_model="model_svm_tfidfvectorizer.tar.gz",
)
resp_count = predictor.predict(
    {"input": ["el servicio fue excelente"]},
    target_model="model_svm_countvectorizer.tar.gz",
)
print(resp_tfidf, resp_count)
```
